```python
import string
fh=open('romeo-full.txt')
d=dict()
for line in fh:
    line=line.rstrip()
    line=line.lower()
    line=line.translate(line.maketrans('','',string.punctuation))
    words=line.split()
    for w in words:
        d[w]=d.get(w,0)+1
fh.close()
d
        
    
```

```python
fh=open('mbox-short.txt')
d=dict()
for line in fh:
    line=line.rstrip()
    if line.startswith("From "):
        words=line.split()
        d[words[2]]=d.get(words[2],0)+1
fh.close()
print(d)
```

```python
#DML
#Insert
import sqlite3
conn=sqlite3.connect('tt.sqlite')
cur=conn.cursor()

title=input('Enter the track title')
plays=int(input('Enter no. of plays'))

cur.execute('INSERT INTO Tracks(title,plays) VALUES(?,?)',(title,plays))
conn.commit() #saves the changes

```

```python
#DML
#Update
import sqlite3
conn=sqlite3.connect('tt.sqlite')
cur=conn.cursor()
title=input('Enter the track title')
plays=int(input('Enter no. of plays'))
cur.execute('update Tracks set plays=? where title=?',(plays,title))
conn.commit() #saves the changes
```

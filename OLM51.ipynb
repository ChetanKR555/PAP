```python
import socket
mysocket=socket.socket(socket.AF_INET,socket.SOCK_STREAM)
mysocket.connect(('data.pr4e.org',80))
cmd='GET http://data.pr4e.org/romeo.txt HTTP/1.0\r\n\r\n'.encode()
mysocket.send(cmd)
while True:
    data=mysocket.recv(512)
    if(len(data)<=0):
        break
    print(data.decode(),end='')
```

```python
import socket
mysocket=socket.socket(socket.AF_INET,socket.SOCK_STREAM)
mysocket.connect(('data.pr4e.org',80))
cmd='GET http://data.pr4e.org/cover3.jpg HTTP/1.0\r\n\r\n'.encode()
mysocket.send(cmd)
content=b''

while True:
    data=mysocket.recv(512)
    if(len(data)<=0):
        break
    content=content+data
pos=content.find(b'\r\n\r\n')
fh=open('lcc.jpg','wb')
fh.write(content[pos+4:])
fh.close()
```

```python
import urllib
fh=urllib.request.urlopen('http://data.pr4e.org/romeo.txt')
for line in fh:
    print(line.decode().strip())
```

```python
import urllib
img=urllib.request.urlopen('http://data.pr4e.org/cover3.jpg')
fh=open('urcc.jpg','wb')
while True:
    chunk=img.read(1024)
    if(len(chunk)<=0):
        break
    fh.write(chunk)
fh.close()
```
